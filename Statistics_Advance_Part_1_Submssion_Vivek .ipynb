```python
#  Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define parameters for the discrete uniform distribution
low = 1  # Minimum value
high = 6  # Maximum value (inclusive)
x_values = np.arange(low, high + 1)  # Possible outcomes

# Compute PMF
pmf_values = stats.randint.pmf(x_values, low, high + 1)

# Plot PMF
plt.figure(figsize=(8, 5))
plt.bar(x_values, pmf_values, color='blue', alpha=0.7, edgecolor='black')

# Labels and title
plt.xlabel("Values")
plt.ylabel("Probability Mass Function (PMF)")
plt.title("Discrete Uniform Distribution (PMF)")
plt.xticks(x_values)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
#  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
import scipy.stats as stats

def bernoulli_pdf(x, p):

    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 in a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")
    
    return stats.bernoulli.pmf(x, p)

# Example usage
p_success = 0.6  # Probability of success
print(f"P(X=1): {bernoulli_pdf(1, p_success):.3f}")
print(f"P(X=0): {bernoulli_pdf(0, p_success):.3f}")
```

```python
# Simulate data from a normal distribution and plot its histogram.
import numpy as np
import matplotlib.pyplot as plt

# Define parameters for the normal distribution
mean = 50  # Mean of the distribution
std_dev = 10  # Standard deviation
size = 1000  # Number of samples

# Generate random samples from a normal distribution
normal_data = np.random.normal(loc=mean, scale=std_dev, size=size)

# Create histogram for visualization
plt.figure(figsize=(8, 5))
plt.hist(normal_data, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')

# Overlay theoretical normal distribution curve
x_values = np.linspace(min(normal_data), max(normal_data), 100)
pdf_values = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_values - mean) / std_dev) ** 2)
plt.plot(x_values, pdf_values, 'r-', label="Theoretical PDF")

# Labels and title
plt.xlabel("Values")
plt.ylabel("Probability Density")
plt.title("Normal Distribution Simulation")
plt.legend()
plt.grid(True)

plt.show()
```

```python
#  Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):

    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Sample standard deviation
    z_scores = (data - mean) / std_dev
    return z_scores

def plot_z_scores(z_scores):
    """
    Plots Z-scores as a histogram to visualize standardization.
    
    Parameters:
    - z_scores (NumPy array): Standardized Z-scores.
    """
    plt.figure(figsize=(8,5))
    plt.hist(z_scores, bins=20, color='blue', alpha=0.7, edgecolor='black', density=True)

    # Overlay standard normal distribution curve
    x_values = np.linspace(min(z_scores), max(z_scores), 100)
    pdf_values = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x_values**2)
    plt.plot(x_values, pdf_values, 'r-', label="Standard Normal PDF")

    # Labels and title
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.title("Z-score Distribution")
    plt.legend()
    plt.grid(True)

    plt.show()

# Example usage
data = np.random.normal(50, 10, 1000)  # Simulated dataset
z_scores = calculate_z_scores(data)
plot_z_scores(z_scores)
```

```python
# Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.
import numpy as np
import matplotlib.pyplot as plt

# Parameters
population_mean = 50
population_std = 15
population_size = 100000
sample_size = 30  # Each sample's size
num_samples = 1000  # Number of samples

# Generate a normally distributed population
population = np.random.normal(population_mean, population_std, population_size)

# Generate multiple sample means
sample_means = [np.mean(np.random.choice(population, size=sample_size, replace=True)) for _ in range(num_samples)]

# Plot histogram of sample means
plt.figure(figsize=(8,5))
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')

# Overlay theoretical normal distribution curve (approximation)
x_values = np.linspace(min(sample_means), max(sample_means), 100)
pdf_values = (1 / (np.std(sample_means) * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_values - np.mean(sample_means)) / np.std(sample_means)) ** 2)
plt.plot(x_values, pdf_values, 'r-', label="Theoretical Normal PDF")

# Labels and title
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.title("Central Limit Theorem Verification")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).
import numpy as np
import matplotlib.pyplot as plt

def standard_normal_distribution():
    """
    Generates and plots the standard normal distribution (mean = 0, std = 1).
    """
    # Generate x-values
    x_values = np.linspace(-4, 4, 1000)

    # Compute PDF for standard normal distribution
    pdf_values = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x_values**2)

    # Plot standard normal distribution curve
    plt.figure(figsize=(8,5))
    plt.plot(x_values, pdf_values, color='blue', label="Standard Normal PDF")
    
    # Highlight mean and standard deviations
    plt.axvline(0, color='black', linestyle='dashed', label="Mean (0)")
    plt.axvline(-1, color='red', linestyle='dashed', label="-1 Std Dev")
    plt.axvline(1, color='red', linestyle='dashed', label="+1 Std Dev")
    
    # Labels and title
    plt.xlabel("Values")
    plt.ylabel("Probability Density")
    plt.title("Standard Normal Distribution (Mean=0, Std=1)")
    plt.legend()
    plt.grid(True)
    
    plt.show()

# Run the function
standard_normal_distribution()
```

```python
# Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal 
# distribution
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_z_score(x, dataset):

    mean = np.mean(dataset)
    std_dev = np.std(dataset, ddof=1)  # Sample standard deviation
    z_score = (x - mean) / std_dev
    return z_score

def plot_z_score(z_score):
    """
    Plots the standard normal distribution and highlights the given Z-score.
    
    Parameters:
    - z_score (float): Calculated Z-score.
    """
    x_values = np.linspace(-4, 4, 1000)
    pdf_values = stats.norm.pdf(x_values, 0, 1)  # Standard normal distribution

    plt.figure(figsize=(8,5))
    plt.plot(x_values, pdf_values, color='blue', label="Standard Normal Distribution")
    plt.axvline(z_score, color='red', linestyle='dashed', label=f"Z-score ({z_score:.2f})")

    # Labels and title
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.title("Z-score Comparison with Standard Normal Distribution")
    plt.legend()
    plt.grid(True)
    
    plt.show()

# Example usage
dataset = np.random.normal(50, 10, 1000)  # Simulated dataset
x_value = 55  # Given data point
z_score = calculate_z_score(x_value, dataset)
print(f"Calculated Z-score for {x_value}: {z_score:.3f}")

plot_z_score(z_score)
```

```python
#  Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
import scipy.stats as stats

# Sample dataset
sample_data = np.random.normal(loc=100, scale=15, size=50)  # Simulated data with mean=100, std=15

# Given population mean (hypothetical)
population_mean = 105

# Compute sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
sample_size = len(sample_data)

# Calculate Z-score
z_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

# Compute p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Interpretation
alpha = 0.05  # Significance level

print(f"Sample Mean: {sample_mean:.3f}")
print(f"Z-statistic: {z_statistic:.3f}")
print(f"P-value: {p_value:.5f}")

if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample and population mean.")
```

```python
#  Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
import scipy.stats as stats

# Simulated dataset
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=50)  # Sample with mean=100, std=15

# Compute sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
sample_size = len(data)

# Confidence level (95%)
confidence_level = 0.95
z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)  # Z-score for 95% confidence

# Calculate margin of error
margin_of_error = z_value * (sample_std / np.sqrt(sample_size))

# Compute confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.3f}")
print(f"Confidence Interval (95%): ({lower_bound:.3f}, {upper_bound:.3f})")
```

```python
# Implement Z-score normalization on a dataset and visualize its transformation.
import numpy as np
import matplotlib.pyplot as plt

# Generate a dataset with a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, Std=15

# Compute Z-score normalized values
mean = np.mean(data)
std_dev = np.std(data, ddof=1)
normalized_data = (data - mean) / std_dev

# Visualization: Before & After Normalization
plt.figure(figsize=(10,5))

# Original Data Histogram
plt.subplot(1,2,1)
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black', density=True)
plt.axvline(mean, color='red', linestyle='dashed', label="Mean (Original)")
plt.xlabel("Values")
plt.ylabel("Density")
plt.title("Original Data Distribution")
plt.legend()
plt.grid(True)

# Normalized Data Histogram
plt.subplot(1,2,2)
plt.hist(normalized_data, bins=30, alpha=0.7, color='green', edgecolor='black', density=True)
plt.axvline(0, color='red', linestyle='dashed', label="Mean (Normalized)")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.title("Z-score Normalized Distribution")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal 
# distribution.
import numpy as np
import scipy.stats as stats

def compute_skewness_kurtosis(data):

    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data, fisher=True)  # Fisher=True returns excess kurtosis
    return skewness, kurtosis

# Generate a dataset from a normal distribution
np.random.seed(42)
normal_data = np.random.normal(loc=100, scale=15, size=1000)

# Compute skewness and kurtosis
skewness, kurtosis = compute_skewness_kurtosis(normal_data)

print(f"Skewness: {skewness:.3f}")
print(f"Kurtosis (Excess): {kurtosis:.3f}")
```
